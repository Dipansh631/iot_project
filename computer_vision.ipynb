# Bitwise Operations on Images

### And

```python
import cv2
import numpy as np

img1=cv2.imread("binary image 1.png")
img2=cv2.imread("binary image 2.png")

re_img1=cv2.resize(img1,(500,500))
re_img2=cv2.resize(img2,(500,500))

new=cv2.bitwise_and(re_img1,re_img2)
h = np.hstack((re_img1,re_img2,new))

cv2.imshow("ghibili",h) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Or

```python
import cv2
import numpy as np

img1 = cv2.imread("binary image 1.png")
img2 = cv2.imread("binary image 2.png")


re_img1 = cv2.resize(img1, (500, 500))  
re_img2 = cv2.resize(img2, (500, 500))  

new = cv2.bitwise_or(re_img1, re_img2)

h = np.hstack((re_img1, re_img2, new)) 

cv2.imshow("ghibili", h)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Rotating

```python
import cv2
# Read the image file
img = cv2.imread("wallpaper.png")

# Resize the image to 500x500 pixels
img1 = cv2.resize(img, (500, 500))

# Get the dimensions of the resized image
# Note: OpenCV stores images as height x width (rows x columns)
w, h = img1.shape[0], img1.shape[1]
print(w, h)

# Create a rotation matrix
# Parameters: center point (x,y), angle of rotation (40 degrees), scale factor (0)
m = cv2.getRotationMatrix2D((w/2, h/2), 40, 0)

# Apply affine transformation (rotation in this case)
# cv2.warpAffine expects dsize as (width, height) tuple
# This is why we pass (h, w) instead of (w, h)
nimg = cv2.warpAffine(img1, m, (h, w))

# Display the rotated image
cv2.imshow("ghibili", nimg) 

# Wait for a key press (0 means wait indefinitely)
cv2.waitKey(0)

# Close all OpenCV windows
cv2.destroyAllWindows()
```

# Image Blurring

### Gaussian Blur

```python
import cv2
import numpy as np

# Read the image file from disk

img = cv2.imread("wallpaper.png")

# Resize the image to 500x500 pixels
# Parameters: source image, target size (width, height)

img1 = cv2.resize(img, (500, 500))

# Apply Gaussian Blur to reduce noise and detail in the image
# Parameters:
# - source image
# - kernel size (must be odd numbers) - larger values create more blur
# - sigma (standard deviation) - 0 means it's calculated from kernel size


g = cv2.GaussianBlur(img1, (117, 117), 0)

# Display the blurred image in a window titled "ghibili"

cv2.imshow("ghibili", g) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Color Filter

```python
import cv2
import numpy as np

vid = cv2.VideoCapture(0)
def xfun(x):
    pass
cv2.namedWindow("demo")
cv2.createTrackbar("lb","demo",0,255,xfun)
cv2.createTrackbar("lg","demo",0,255,xfun)
cv2.createTrackbar("lr","demo",0,255,xfun)

cv2.createTrackbar("ub","demo",0,255,xfun)
cv2.createTrackbar("ug","demo",0,255,xfun)
cv2.createTrackbar("ur","demo",0,255,xfun)





while vid.isOpened():
    r,frame=vid.read()
    if r==True:
        img=cv2.resize(frame,(400,400))
        img=cv2.flip(img,1)
        hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
        lb=cv2.getTrackbarPos("lb","demo")
        lg=cv2.getTrackbarPos("lg","demo")
        lr=cv2.getTrackbarPos("lr","demo")
        
        ub=cv2.getTrackbarPos("ub","demo")
        ug=cv2.getTrackbarPos("ug","demo")
        ur=cv2.getTrackbarPos("ur","demo")

        lo=np.array([lb,lg,lr])
        up=np.array([ub,ug,ur])

        masks=cv2.inRange(hsv,lo,up)

        res=cv2.bitwise_and(img,img,mask=masks)
        
        cv2.imshow('res',res)
        cv2.imshow('masks',masks)
        cv2.imshow('hsv',hsv)
        cv2.imshow('hulooloo',img)

        if cv2.waitKey(10) & 0xff==ord("p"):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()
```

# Object Detection in vedios

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


cap=cv2.VideoCapture(r"C:\Users\MSI1\OneDrive\Desktop\Untitled video - Made with Clipchamp.mp4")
r,f=cap.read()
x,y,w,h=250,220,400,400
t=(x,y,w,h)
roi=f[y:y+h,x:x+w]
hsv_roi=cv2.cvtColor(roi,cv2.COLOR_BGR2HSV)
mask=cv2.inRange(hsv_roi,np.array((0.,60.,32.)),np.array((180.,255.,255.)))
hist=cv2.calcHist([hsv_roi],[0],mask,[180],[0,180])
cv2.normalize(hist,hist,0,255,cv2.NORM_MINMAX)
ter=(cv2.TERM_CRITERIA_EPS  | cv2.TERM_CRITERIA_COUNT,10,1)





while cap.isOpened():
    r,f=cap.read()
    if  r==True:
        hsvf=cv2.cvtColor(f,cv2.COLOR_BGR2HSV)
        d=cv2.calcBackProject([hsvf],[0],hist,[0,180],1)
        r,tp=cv2.meanShift(d,t,ter)
        x,y,w,h=tp
        final=cv2.rectangle(f,(x,y),(x+w,y+h),(0,0,255),4)
        cv2.imshow("ghibili",final)
        if cv2.waitKey(15) & 0xff==ord("x"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

# Face Detection

```python
import cv2
import mediapipe as mp
```

```python
mp_face=mp.solutions.face_detection
mp_draw=mp.solutions.drawing_utils
```

```python
face_dect=mp_face.FaceDetection(min_detection_confidence=1,model_selection=0)
vid = cv2.VideoCapture(0)
while True:
    r,frame=vid.read()
    frame=cv2.flip(frame,1)
    if r==True:
        frame=cv2.resize(frame,(700,500))
        result=face_dect.process(frame)
        if r==True:
            for cr in result.detections:
                mp_draw.draw_detection(frame,cr)
            
        cv2.imshow("h",frame)
        if cv2.waitKey(45) & 0xff==ord(" "):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()
```

# Hand Tracking

```python
import cv2
import mediapipe as mp
mp_draw=mp.solutions.drawing_utils
mp_drawing_styles=mp.solutions.drawing_styles
mp_hands=mp.solutions.hands
p=mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.5)
vid = cv2.VideoCapture(0)
while True:
    r,frame=vid.read()
    frame=cv2.flip(frame,1)
    if r==True:
        frame=cv2.resize(frame,(700,500))
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert to RGB
        result=p.process(rgb_frame)
        if result.multi_hand_landmarks:
            for cr in result.multi_hand_landmarks:
                mp_draw.draw_landmarks(frame,cr,mp_hands.HAND_CONNECTIONS)
            
        cv2.imshow("h",frame)
        if cv2.waitKey(45) & 0xff==ord(" "):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

mp_draw = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

p = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)

vid = cv2.VideoCapture(0)

while True:
    r, frame = vid.read()
    frame = cv2.flip(frame,1)

    if r==True:
        frame = cv2.resize(frame, (700, 500))
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert to RGB

        result = p.process(rgb_frame)

        if result.multi_hand_landmarks:
            for cr in result.multi_hand_landmarks:
                mp_draw.draw_landmarks(frame, cr, mp_hands.HAND_CONNECTIONS)

        cv2.imshow("Hand Detection", frame)

        if cv2.waitKey(45) & 0xFF == ord(" "):
            break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```
